### DDA 분석 결과

- 변수 선택
    - 목적 변수 : '입원기간'
    - 설명변수 선택 : '통증기간(월)','연령','수술시간','디스크단면적','과거수술횟수','환자통증정도'

- 변수 선택 기준

    - 입원기간에 영향이 있을 것 같은 변수들을 선택
    - '연령' : 나이에 따라서 회복 경과가 어떻게 되는지 궁금함.
    - '수술시간' : 수술시간이 길어지면 입원기간이 길어지는지 파악. 
    - '과거수술횟수' : 과거에 수술했던 경험이 있으면 입원기간이 길어지는 확인하고자 함
    - '디스크단면적': 해당 크기가 입원기간에 어떤 영향을 끼치는지 확인하고자 함.
    
- DDA 분석 결과
    입원기간의 최장기간은 51일, 최단기간은 0일임. 입원기간의 평균값은 1.73일, 중앙값은 1임.
 

```python
import matplotlib.pyplot as plt
import matplotlib as mpl 
mpl.rc('font',family='Malgun Gothic')
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df_ROS = pd.read_csv('../datasets/RecurrenceOfSurgery.csv')
df_ROS.head()
```

```python
df_ROS_data = df_ROS[['입원기간','통증기간(월)','연령','수술시간','디스크단면적','과거수술횟수','환자통증정도']]
df_ROS_data
```

#### 데이터 값을 보아 '과거수술횟수','환자통증정도','입원기간' 는 object 변환 해야 할 듯

```python
df_ROS_data[['과거수술횟수','환자통증정도','입원기간']] = df_ROS_data[['과거수술횟수','환자통증정도','입원기간']].astype(object)
df_ROS_data.info()
```

```python
pd.pivot_table(data= df_ROS_data, index='입원기간', columns=['통증기간(월)','연령','수술시간','디스크단면적','과거수술횟수','환자통증정도'], aggfunc='size')
```
