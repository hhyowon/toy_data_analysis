### BMI , BMI_카테고리 컬럼 추가

```python
def calculate_bmi(체중, 신장):
    return 체중 / (신장 / 100) ** 2
```

```python
df_ROS_Q['BMI'] = df_ROS_Q.apply(lambda row: calculate_bmi(row['체중'], row['신장']), axis=1)
```

```python
def bmi_categori(bmi):
    if bmi < 18.5:
        return "저체중"
    elif 18.5 <= bmi < 24.9:
        return "정상 체중"
    elif 25 <= bmi < 29.9:
        return "과체중"
    else:
        return "비만"
```

```python
df_ROS_Q['BMI_categori'] = df_ROS_Q['BMI'].apply(bmi_categori)
```

```python
df_ROS_BMI = df_ROS_Q[['환자ID','신장','체중', 'BMI','BMI_categori','입원기간']]
df_ROS_BMI
```
